```python
import pandas as pd
```

```python
data = pd.read_csv("cancer dataset.csv")
```

```python
data['diagnosis'].value_counts()
```

```python
data.columns
```

```python
data
```

```python
cat_datatype = []
num_datatype = []
for col in data.columns:
    if data[col].dtype == 'O':
        cat_datatype.append(col)
    else:
        num_datatype.append(col)
```

```python
print("Columns with Object datatype: ") 
cat_datatype
```

```python
print("Columns with Non-object Datatype: ")
num_datatype
```

```python
data['diagnosis'].value_counts()
```

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 
```

```python
data['diagnosis']= label_encoder.fit_transform(data['diagnosis']) 
```

```python
data['diagnosis'].value_counts()
```

```python
notnul_cols = []
for col in data.columns:
    if data['diagnosis'].isnull().sum() != 0 :
        notnul_cols.append(col)
```

```python
print("Number of columns with missing data: ",len(notnul_cols))
```

```python
data.corr()['diagnosis']
```

```python
less_corr_cols = []

for col in data.columns:
    if abs(data.corr()[col]['diagnosis']) < 0.3:
        less_corr_cols.append(col)
```

```python
print("Columns with correlation value less than +/- 0.3 are : ")
less_corr_cols
```

```python
new_data = data.drop(columns= less_corr_cols, axis = 1)
```

```python
new_data
```

```python
attributes = new_data.drop(columns = ['diagnosis'], axis=1).values
```

```python
attributes
```

```python
target = new_data['diagnosis']
```

```python
target
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
attributes = sc.fit_transform(attributes)
```

```python
attributes
```

```python
from sklearn.model_selection import train_test_split 
XTrain, XTest, YTrain, YTest = train_test_split(attributes, target, test_size=0.20, shuffle=False)
```

```python
print("Shape of Training dataset: ")
print(XTrain.shape)
```

```python
print("Shape of Testing dataset: ")
print(XTest.shape)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(XTrain, YTrain)
```

```python
log_pred = log_model.predict(XTest)
log_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
cnf_matrix = confusion_matrix(YTest, log_pred)
print("Confusion matrix")
print(cnf_matrix)

print("Performance of the Logistic Regression Model : ")
print(f"Accuracy: {accuracy_score(YTest, log_pred) *100 :0.3f}")
print(f"Precision: {precision_score(YTest, log_pred) *100 :0.3f}")
print(f"Recall: {recall_score(YTest, log_pred) *100 :0.3f}")
print(f"F1 Score: {f1_score(YTest, log_pred) *100 :0.3f}")
```

```python
from sklearn.svm import SVC
svm_clf = SVC()
svm_clf.fit(XTrain, YTrain)
```

```python
svm_pred = svm_clf.predict(XTest)
svm_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
cnf_matrix = confusion_matrix(YTest, svm_pred)
print("Confusion matrix")
print(cnf_matrix)

print("Performance of the SVM Classifier : ")
print(f"Accuracy: {accuracy_score(YTest, svm_pred) *100 :0.3f}")
print(f"Precision: {precision_score(YTest, svm_pred) *100 :0.3f}")
print(f"Recall: {recall_score(YTest, svm_pred) *100 :0.3f}")
print(f"F1 Score: {f1_score(YTest, svm_pred) *100 :0.3f}")
```

```python
from sklearn.ensemble import RandomForestClassifier
random_clf = RandomForestClassifier(n_estimators=100, random_state=42)
random_clf.fit(XTrain, YTrain)
```

```python
rf_pred = random_clf.predict(XTest)
rf_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
cnf_matrix = confusion_matrix(YTest, rf_pred)
print("Confusion matrix")
print(cnf_matrix)

print("Performance of the Random Forest Classifier : ")
print(f"Accuracy: {accuracy_score(YTest, rf_pred) *100 :0.3f}")
print(f"Precision: {precision_score(YTest, rf_pred) *100 :0.3f}")
print(f"Recall: {recall_score(YTest, rf_pred) *100 :0.3f}")
print(f"F1 Score: {f1_score(YTest, rf_pred) *100 :0.3f}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense 
from tensorflow.keras.optimizers import Adam

model = Sequential([
    Dense(32, activation='relu', input_shape=(23,)),
    Dense(1, activation='sigmoid')  
])

model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(XTrain, YTrain, epochs=10, batch_size=8, validation_split=0.2)
```

```python
model.evaluate(XTest, YTest)
```

```python
ann_pred = model.predict(XTest)
ann_pred
```
